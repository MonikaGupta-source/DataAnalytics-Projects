```python
sample_df['burnout_score'].describe()
```

```python

percentile_1 = np.percentile(sample_df['burnout_score'], 25)
percentile_3 = np.percentile(sample_df['burnout_score'], 75)
percentile_2 = np.percentile(sample_df['burnout_score'], 50)

IQR = percentile_3 - percentile_1

print("Q1:", percentile_1)
print("Q3:", percentile_3)
print("Q2:", percentile_2)
print("IQR:", IQR)

```

```python
sns.boxplot(y=sample_df['burnout_score'])
```

```python

upper_bound = percentile_3 + 1.5 * IQR
lower_bound = percentile_1 - 1.5 * IQR

print("The upper fence outlier : ",upper_bound)
print("The lower fence outlier : ",lower_bound)
```

```python
outliers = sample_df[sample_df['burnout_score']>upper_bound]['burnout_score']
outliers
```

**PROBLEM IN THE DATA**

In our burnout_score data we indentified 2 outliers at values 7.22 and 7.30 these values are significatly higher than pur upper fence of 6.79

**WHY THIS IS AN PROBLEM**

These outliers were distorting our mean and standard deviation, making our overall analysis less accurate. For example , our actual mean without outliers will be lower and more representative of majority students

**LARGE IMPACT OF THE OUTLIERS**

In machine learning models , only 2 outliers could disproportionately influence our predictions leading to poor generalization which means our model would perform well on these 2 items but fail on the reamining 99 items

**ACTION SHOULD BE PERFROMED**

We have to removed these 2 outliers and our clean dataset now represent the true pattern of student burnout more accurately

```python
# remove outliers
clean_df = sample_df[sample_df['burnout_score'] <= upper_bound]

# verify
print("Before removing outliers:", len(sample_df))
print("After removing outliers:", len(clean_df))
```

```python
print("BEFORE:")
print(sample_df['burnout_score'].describe())

print("\nAFTER:")
print(clean_df['burnout_score'].describe())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# before
sns.boxplot(y=sample_df['burnout_score'], ax=axes[0])
axes[0].set_title('Before Removing Outliers')

# after
sns.boxplot(y=clean_df['burnout_score'], ax=axes[1])
axes[1].set_title('After Removing Outliers')

plt.suptitle('Burnout Score — Outlier Treatment', fontsize=14)
plt.show()
```

**conclusion**

we have indentified burnou cases(7.22 and 7.30) beyond our upper fence of 6.79 removing them giving us a cleaner more accurate picture of sudent burnout and ensures our fuere predictiom are not biased by these extreme cases.

**HISTPLOT | FREQUENCY POLYGON | KDE | ALL IN ONE FRAME**

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Histogram
sns.histplot(x=sample_df['sleep_hours'], ax=axes[0])
axes[0].set_title('Histogram')

# 2. Frequency Polygon
counts, bins = np.histogram(sample_df['sleep_hours'], bins=10)
bin_centers = (bins[:-1] + bins[1:]) / 2
axes[1].plot(bin_centers, counts, marker='o', color='blue')
axes[1].set_title('Frequency Polygon')
axes[1].set_xlabel('sleep_hours')
axes[1].set_ylabel('Frequency')

# 3. KDE
sns.kdeplot(x=sample_df['sleep_hours'], ax=axes[2])
axes[2].set_title('KDE Plot')

plt.suptitle('Sleep Hours — All 3 Plots', fontsize=14)
plt.show()
```

I created all 3 visualization side by side to show the progression from raw data to samooth curve and to prove my findings are consitent across all the methods.HISTOGRAM shows exact bars RAW AND ACCURATE........... FREQUENCY POLYGON connects bar tops with lines the dip between 4-5 hours and sharp peak at 6 hours very visible here and a the last KDE(KERNEL DENSITY ESTIMATOR) show the smoothest view removes all noise best for non technical audience The small bump at 3-4 is visible here whcih hints at a seperate group of the sleep - deprived students

Q8: Plot a pairplot of [‘sleep_hours’, ‘physical_activity’, ‘depression_score’, ‘mental_health_index’]. Which pair shows the strongest linear trend?

```python
# Get the correlation matrix
cols = ['sleep_hours', 'physical_activity', 'depression_score', 'mental_health_index']
correlation_matrix = sample_df[cols].corr()
print("Correlation Matrix:\n", correlation_matrix)
```

CORRELATION IS ALWAYS IN THE -1 TO +1 RANGE : 

Range             
-1.0 to -0.7 --> Strong Negative
-0.7 to -0.4 --> Moderate Negative
-0.4 to 0.0 -->  Weak Negative
0         --->   No Relation
0.0 to 0.4--->   Weak Positive
0.4 to 0.7---    Moderate Positive
0.7 to 1.0--->   Strong Positive

**HEXBIN PLOT**

Q9: Create a hexbin plot of screen_time vs anxiety_score (useful for large data). Where is the highest density of points?

```python
plt.figure(figsize=(8,6))
plt.hexbin(x=sample_df['screen_time'], 
           y=sample_df['anxiety_score'], 
           gridsize=20, 
           cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Screen Time')
plt.ylabel('Anxiety Score')
plt.title('Hexbin Plot - Screen Time vs Anxiety Score')
plt.show()
```

Instead of many overlapping points (overplotting), the space is divided into hexagons.
Colour (or count) shows how many data points fall inside each hexagon.
Darker/warmer colour → higher density.
