**PERFORMING A HYPOTHESIS TESTING**

```python
from scipy.stats import stats
```

**SAMPLE DATA**

```python
sample_df = df.loc[0:100,:]
sample_df
```

Question: Is there a significant difference in depression_score between male and female students? 

**PERFORMING T-TEST [TWO SAMPLE T-TEST]**

According to the formula of the t-test we want:
i. sample mean 
ii. sample size 
iii. population mean 
iv. population std 

```python
sample_df.dtypes
```

we have to extract the males student and there corresponing dipression score and same goes with female students 

```python
male_group = sample_df[sample_df['gender']=='Male']['depression_score']
male_group
```

```python
female_group = sample_df[sample_df['gender']=='Female']['depression_score']
female_group
```

```python
m_mean = male_group.mean()
std_m = male_group.std()
n_m = len(male_group)
print(f"Male: mean={m_mean:.2f}, std={std_m:.2f}, n={n_m}")
```

```python
f_mean = female_group.mean()
std_f = female_group.std()
n_f = len(female_group)
print(f"Female: mean={f_mean:.2f}, std={std_f:.2f}, n={n_f}")
```

**APPLING THE FORMULA (manual method)**

```python
t_stat = (m_mean - f_mean) / np.sqrt((std_m**2 / n_m) + (std_f**2 / n_f))
t_stat
```

```python
df = ( (std_m**2/n_m + std_f**2/n_f)**2 ) / ( (std_m**2/n_m)**2/(n_m-1) +
        (std_f**2/n_f)**2/(n_f-1) )
df
```

```python
from scipy.stats import t
```

```python
p_value = 2 * (1 - t.cdf(abs(t_stat), df))
p_value
```

**STEP 4 : DRAWING CONCLUSION**

```python
alpha = 0.05
```

```python
if p_value< alpha:
    print("Rejecting the H₀.")
else:
    print("Fail to reject H₀.")
```

 **DIRECT METHOD(using ttest_ind)**

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(male_group, female_group)
print(f"t = {t_stat:.5f},p-value = {p_value:.5f}")
```

```python
alpha = 0.05
if p_value< alpha:
    print("Rejecting the H₀.")
else:
    print("Fail to reject H₀.")
```
